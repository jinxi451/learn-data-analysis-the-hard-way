### 查看data_train 中 'Survived' 的按照'Sex', 的数量分布data_group(groupby)

```python
# 两个变量关系
data_group = data_train[['Sex','Survived']].groupby(['Sex','Survived']).size()
```

```python
data_group.to_frame('num')
```

### 将聚合后的数据绘图

```python
data_group.plot(kind ='bar')
```

```python
data_train.groupby(['Sex','Survived'])['Survived'].count()
```

### 查看data_train 中 'Survived' 的按照'Title',统计'mean','count','sum'(groupby)

```python
data_train[['Title', 'Survived']].groupby(['Title'],as_index =False).agg(['mean','count','sum'])
```

##  数据透视表 ： pivot_table  (数据透视图)

### 对某一变量求和,以Title为行 Survived为列，以Family_Size为值： len(可行) /count

```python
pivot1 = pd.pivot_table(data_train,index=['Title'],columns=['Survived'],values=['Family_Size'], aggfunc=np.sum )
pivot1
```

### 对某一变量求和,以Title为行 Survived为列，以Family_Size为值，计算样本数、和、均值

```python
pivot2 = pd.pivot_table(data_train,index=['Title'], \
                    columns=['Survived'], \
                    values=['Family_Size'], \
                    aggfunc=[len,np.sum,np.mean] )
pivot2
```

### 对透视表进行切片
#首先进行 切片 然后再进行 透视表

```python
pivot1[pivot1.sum(axis = 1) >400]
```

### MultiIndex索引

```python
pivot2 = pd.pivot_table(data_train,index=['Title','Sex'], \
                    columns=['Embarked','Survived'], \
                    values=['Family_Size'], \
                    aggfunc=[len,np.sum,np.mean] )
pivot2
```

### 依据行 选取Title 为Master和Officer 的透视数据

```python
pivot2.iloc[pivot2.index.get_level_values('Title').isin(['Master','Officer']),:]
```

### 依据行 选取列 为sum 的透视数据

```python
pivot2.iloc[:,pivot2.columns.get_level_values(0)=='sum']
```

### 基于query选取Title 为Master 的数据

```python
pivot2.query('Title == "Master"')
```

### 统计Survived 与 Title2 之间频数分布关系

#### 仅有两列时 怎么看 数量 ，和 每个的百分比. 频数 crosstab() 

```python
pd.crosstab(data_train['Survived'],data_train['Title'],margins =True)
```

### 在corsstab 中显示百分比

### 按行 求总体占比

```python
pd.crosstab(data_train['Survived'], data_train['Title']).apply(lambda r: r/r.sum(), axis=1)  
```

### 按列 求总体占比

```python
pd.crosstab(data_train['Survived'], data_train['Title']).apply(lambda r: r/r.sum(), axis=0)  
```

### 按全部 求总体占比

```python
pd.crosstab(data_train['Survived'], data_train['Title']).apply(lambda r: r/len(data_train))
```

```python
#只返回2个变量的计数
def table(data,col_list = None,margins = False):
    return pd.pivot_table(data[col_list],index=col_list[0],columns=col_list[1], aggfunc=[len] ,margins=margins)

table(data_train,['Survived','Title'])
```
